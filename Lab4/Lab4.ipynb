```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import data, exposure

# Load the moon image
img = data.moon()

# Contrast stretching: Rescale intensity to the 3rd and 80th percentiles
p3, p80 = np.percentile(img, (3, 80))
img_rescale = exposure.rescale_intensity(img, in_range=(p3, p80))

# Display the image and its histogram
fig = plt.figure(figsize=(10, 7))

# Plot the rescaled image
fig.add_subplot(2, 1, 1)
plt.imshow(img_rescale, cmap='gray')
plt.axis('off')
plt.title('Contrast stretch (3rd to 80th percentile)')

# Plot the histogram
fig.add_subplot(2, 1, 2)
plt.hist(img_rescale.flat, bins=256, range=(0, 255))
plt.xlabel('gray level values')
plt.ylabel('Number of pixels')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from skimage import data, exposure

# Load the moon image
img = data.moon()

# Flatten the histogram using equalize_hist
img_eq = exposure.equalize_hist(img)

# Display the image and its equalized histogram
fig = plt.figure(figsize=(10, 7))

# Plot the equalized image
fig.add_subplot(2, 1, 1)
plt.imshow(img_eq, cmap='gray')
plt.axis('off')
plt.title('Equalized Histogram Image')

# Plot the histogram
fig.add_subplot(2, 1, 2)
plt.hist(img_eq.flat, bins=256, range=(0, 1.0))
plt.xlabel('gray level values (normalized)')
plt.ylabel('Number of pixels')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from skimage import data
from skimage.exposure import match_histograms

# Load the images from skimage.data
reference = data.rocket()
image = data.chelsea()

# Implement histogram matching
# channel_axis=-1 ensures it applies to color channels correctly
matched = match_histograms(image, reference, channel_axis=-1)

# Display the results side-by-side
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

axes[0].imshow(image)
axes[0].set_title('Source Image (Chelsea)')
axes[0].axis('off')

axes[1].imshow(reference)
axes[1].set_title('Reference Image (Rocket)')
axes[1].axis('off')

axes[2].imshow(matched)
axes[2].set_title('Matched Image')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
